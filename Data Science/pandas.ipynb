# 1 Creating dataframe from an array
### 1.1 Option 1

```python
# creating an array
data=np.array([[1,4],[2,5],[3,6]])
```

# Creating Datframes from a dictionary

```python
# for example
states=['Uttar Pradesh','Rajasthan','Karnatka','Tamil Nadu','Maharashtra']
popultion=[2343535,3464332,1345434,1436345,2343242]
```

```python
#storing list
dict_states={'States':states,"Population":popultion}
```

```python
#Creating data frame
df_population=pd.DataFrame(dict_states,index=[1,2,3,4,5])
```

```python
#show dataframe
df_population
```
