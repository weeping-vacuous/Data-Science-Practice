# Filtering 

```python
import pandas as pd
```

```python
df_laptops=pd.read_csv('SampleDocuments\\laptop_price.csv',encoding='latin1')
df_laptops.head(3)
```

```python
# find which rows have the word apple in the xcompany name
df_laptops['Company']=='Apple'
```

```python
#filter the data based on the condition
df_laptops[df_laptops['Company']=='Apple']
```

```python
# varify 
df_laptops[df_laptops['Company']=='Apple'].value_counts('Company')
```

```python
# find which rows dont have hp in the company column
df_laptops['Company']!='HP'
```

```python
# filter based on the condition
df_laptops[df_laptops['Company']!='HP']
```

## 1.2 Exercise 2

```python
# find out which laptops have prices over 2000 euros
df_laptops[df_laptops['Price_euros']>2000]
```

# Filtering based on multiple conditons

```python
# find apple laptops that cost more than 2000 euros
(df_laptops['Company']=="Apple") & (df_laptops['Price_euros']>2000)
```

```python
df_laptops[(df_laptops['Company']=="Apple") & (df_laptops['Price_euros']>2000)]
```

```python
# find laptops or dell laptops
(df_laptops['Company']=='Apple')|(df_laptops['Company']=='Dell')
```

```python
df_laptops[(df_laptops['Company']=='Apple')|(df_laptops['Company']=='Dell')]
```

```python
# find laptops from apple or dell that cost more than 2000 euros
df_laptops[((df_laptops['Company']=='Apple')|(df_laptops['Company']=='Dell'))&(df_laptops['Price_euros']>2000)]
```
