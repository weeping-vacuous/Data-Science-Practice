```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# DATA Visualization

```python
df= pd.read_csv(r'D:\Data Science\SampleDocuments\population_total.csv',encoding='latin1')
df
```

### Creating a pivot table

```python
df.dropna(inplace=True)
```

```python
df_pivot=df.pivot_table(index="year",columns="country",values="population")
df_pivot.head(10)
```

```python
df_pivot=df_pivot[['United States','India','China','Indonesia','Brazil']]
```

```python
df_pivot.head(10)
```

```python
plt.figure(figsize=(10,6))
for col in df_pivot.columns:
    plt.plot(df_pivot.index,df_pivot[col],label=col)
plt.title("Populations of Nations")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.tight_layout()
plt.show()
```

## Barplot

```python
# select one year
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_2020=df_pivot_2020.T
df_pivot_2020
```

```python
x_values = df_pivot_2020.index
y_values = df_pivot_2020.iloc[:,0]
plt.bar(x_values,y_values,width = 0.55,color="orange")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# barplot group by n-variables
df_pivot_y=df_pivot[df_pivot.index.isin([1980,1990,2000,2010,2020])]
df_pivot_y
```

```python

y_values_US = df_pivot_y.iloc[:,0]
x_values = np.arange(len(df_pivot_y.index))
width=0.12
plt.bar(x_values-2*width,y_values_US,color="orange",label="United States",width=width)
y_values_IN = df_pivot_y.iloc[:,1]
plt.bar(x_values-width,y_values_IN,color="blue",label="India",width=width)
y_values_CN = df_pivot_y.iloc[:,2]
plt.bar(x_values,y_values_CN,color="red",label="China",width=width)
y_values_IND = df_pivot_y.iloc[:,3]
plt.bar(x_values+width,y_values_IND,color="green",label="Indonesia",width=width)
y_values_BR = df_pivot_y.iloc[:,4]
plt.bar(x_values+2*width,y_values_BR,color="yellow",label="Brazil",width=width)
plt.xticks(ticks=x_values,labels=df_pivot_y.index)
plt.legend()
plt.xlabel("Years")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```
