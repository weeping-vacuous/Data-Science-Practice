Task: Find the cheapest and most expensive of each laptop of each company using sort_values() and duplicated() methods (keep 'first','Last', and false)

```python
# sort dataframe ascending by company and price
# cheapest price first and expensive last 
df_laptops.sort_values(['Company','Price_euros'],inplace=True)
df_laptops
```

```python
#check all the valus in "company"
df_laptops.value_counts('Company')
```

```python
# duplicated values inside company comlumn using keep=first
duplicated_first=df_laptops.duplicated('Company',keep='first')
```

```python
# show data frame with the duplicated values in 'Company column 
df_laptops[duplicated_first]
# show data fraame with non-duplicate values
# since it is already sorted by cheapest to expensive  
df_laptops['Cheapest']=np.where(~duplicated_first,df_laptops['Price_euros'],'-')

```

```python
# Doing same for the most expensive laptops
duplicated_last=df_laptops.duplicated('Company',keep='last')
df_laptops['Most_expensive']=np.where(~duplicated_last,df_laptops['Price_euros'],'-')
```

```python
# Answer
df_laptops[(~duplicated_first)|(~duplicated_last)][['Company','Cheapest','Most_expensive']]
```

# Drpopping duplicated values

```python
# reset dataframe
df_laptops.drop(columns=['Cheapest','Most_expensive'],axis=1,inplace=True)
```

```python
df_laptops.head(3)
```

```python
# drop duplicates in 2 or more columns
df_laptops.drop_duplicates(['Company']).value_counts('Company')
```

```python
df_laptops.drop_duplicates(['Company'],keep='last',ignore_index=True)[['Company','Price_euros']]
```
