### 1.3 With assign()


  i) Add multiple columns in a single line of code  
  ii) When you need to overwrite an existing column

```python
score1=np.random.randint(30,100,size=1000)
score2=np.random.randint(30,100,size=1000)
```

```python
#crating series to add
series1=pd.Series(score1,index=np.arange(0,1000))
series2=pd.Series(score2,index=np.arange(0,1000))
```

```python
series1
```

```python
df_data=df_data.assign(Score1=series1,Score2=series2)
df_data
```

### 1.3 Insert()

Inserts a new column at a specific position

```python
df_data.insert(1,'Test',series1)
```
