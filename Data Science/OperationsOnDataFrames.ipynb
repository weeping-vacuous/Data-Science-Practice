```python
#calculate the sum
df_data['Score']+df_data['Age']
```

```python
score1=np.random.randint(30,100,size=1000)
score2=np.random.randint(30,100,size=1000)
#crating series to add
series1=pd.Series(score1,index=np.arange(0,1000))
series2=pd.Series(score2,index=np.arange(0,1000))
df_data=df_data.assign(Chemistry=series1,Physics=series2)
df_data
```

```python
# Average of a row as a new column
df_data['Average Score']=(df_data['Score']+df_data['Physics']+df_data['Chemistry'])/3
```
