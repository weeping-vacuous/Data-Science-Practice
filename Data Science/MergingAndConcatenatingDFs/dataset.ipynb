```python
import pandas as pd
```

```python
df_movies = pd.read_csv(r'D:\Data Science\SampleDocuments\IMDb movies.csv',low_memory=False)
```

# Concatenate()
1. Vertical

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['E', 'F', 'G', 'F'],
                    'age': [40, 21, 19, 24]})
```

```python
#COncatenating
pd.concat([df1,df2],axis = 0,ignore_index = True)
```

```python
# extract 50% sample of the original dataframe
df_sample = df_movies.sample(frac=.5,random_state=1)
```

```python
# shape od DFs we concatenate
df_sample.shape,df_movies.shape
```

```python
# concatenatnating df_sample and df_movies
df_concat = pd.concat([df_movies,df_sample],axis = 0)
```

```python
# shape
df_concat.shape
```

### Concatenating horizontly

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'job': ['Doctor', 'Statistician',
                            'Accountant', 'Developer']})
```

```python
pd.concat([df1,df2],axis=1)
```

```python
# shapes of dataframes that well concatenate
df_movies.shape,df_ratings.shape
```

# .merge()

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['C', 'D', 'E', 'F'],
                    'job': ['Doctor', 'Statistician',
                            'Accountant', 'Developer']})
```

```python
df1
```

```python
df2
```

```python
# inner join
df_inner_join = df1.merge(df2,on='id',how='inner')
```

```python
df1.shape,df2.shape,df_inner_join.shape
```

```python
# merge of df_movies and df_ratings
df_movies.merge(df_ratings,on='imdb_title_id',how='inner')
```
