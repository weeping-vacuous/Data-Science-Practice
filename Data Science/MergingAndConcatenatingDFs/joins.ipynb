```python
import pandas as pd
```

```python
df_movies = pd.read_csv(r'D:\Data Science\SampleDocuments\IMDb movies.csv',low_memory=False)
df_ratings = pd.read_csv(r'D:\Data Science\SampleDocuments\IMDb ratings.csv',low_memory=False)
df_movies = df_movies[['imdb_title_id','title','year','genre','country']]
df_ratings = df_ratings[['imdb_title_id','total_votes','mean_vote']]
```

# .merge()

Inner join

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['C', 'D', 'E', 'F'],
                    'job': ['Doctor', 'Statistician',
                            'Accountant', 'Developer']})
```

```python
df_inner_join = df1.merge(df2,on='id',how='inner')
```

```python
# merge of df_movies and df_ratings
df_movies.merge(df_ratings,on='imdb_title_id',how='inner')
```

Outer join

```python
df1.merge(df2,on='id',how='outer')
```

```python
# merge of df_movies and df_ratings
df_movies.merge(df_ratings,on='imdb_title_id',how='outer')
```

Exclusive full join

```python
df1.merge(df2,on='id',how='outer',indicator=True)
```

```python
df1.merge(df2,on='id',how='outer',indicator=True).query("_merge == 'left_only' or _merge == 'right_only'")
```

```python
# make a copy of the df_movies
df_copy = df_movies.copy(deep=False) 
```

```python
df_copy.iloc[:1000,0] = 'tt1234567890'
```

```python
# merge of df copy and df rating exclusive left
df_copy.merge(df_ratings,on='imdb_title_id',how='left',indicator=True).query("_merge == 'left_only'")
```

```python
# merge of df copy and df rating exclusive right
df_copy.merge(df_ratings,on='imdb_title_id',how='right',indicator=True).query("_merge == 'right_only'")
```
