```python
import pandas as pd
```

```python
df=pd.read_csv('players_20.csv')
df.set_index('short_name',inplace=True)
df=df[['long_name'	,'age',	'dob',	'height_cm'	,'weight_kg',	'nationality',	'club']]
```

```python
df.head(6)
```

# 1. Set value to one cell

```python
# update L. Messi
df.loc['L. Messi','height_cm'] = 175
```

```python
df
```

# 2. Set value to entire column

```python
# set height of all players to 190
df.loc[:,'height_cm']=190
df
df.drop('height_cm',inplace=True)
```

# 3. Set value to an entire row

```python
# get all columns that correspond to player ranked last in fifa
df.iloc[-1,:]
```

```python
# set null value to all columns that carrespond to player ranked last
import numpy as np
df.iloc[-1,:]=np.nan
```

```python
df.iloc[-1]
```

# 4. Set value for multiple rows

```python
#, Set value for all items matching the list of labels
df.loc[['L. Messi','Cristiano Ronaldo'],['height_cm']]=[175,170]
df
```

# 5. Set value for rows matching a condition

```python
# set value for rows matvhing a condition
df.loc[df['height_cm']>180,'height_cm']=175
df
```
