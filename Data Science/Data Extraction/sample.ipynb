```python
import pandas as pd
```

```python
df=pd.read_csv('players_20.csv')
df.set_index('short_name',inplace=True)
df=df[['long_name'	,'age',	'dob',	'height_cm'	,'weight_kg',	'nationality',	'club']]
```

```python
# exract 10 random elements from nationality column
df['nationality'].sample(10,random_state=99)
```

```python
# 20% sample of data
df.sample(frac=.2,random_state=0)
```

```python
# increasing sampling rate
df.sample(replace=True,frac=2)
```
