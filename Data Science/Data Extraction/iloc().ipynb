```python
import pandas as pd
```

```python
df=pd.read_csv('players_20.csv')
df.set_index('short_name',inplace=True)
df=df[['long_name'	,'age',	'dob',	'height_cm'	,'weight_kg',	'nationality',	'club']]
```

```python
df.head(10)
```

# 1. Selecting with a single value

```python
#height of lionel messi
df.iloc[0,3]
```

```python
# weight of critiano ronaldo
df.iloc[1,4]
```

```python
# get all rows inside the 'height_cm' column
df.iloc[:,3]
```

```python
# get all columns to L. Messi
df.iloc[0,:]
```

# Selecting with a list of values

```python
# get all data about L.Messi and christiano ronaldo 
df.iloc[[0,1],:]
```

```python
# get only height of these players
df.iloc[[0,1],3]
```

```python
# get height and weight if lionel messi
df.iloc[[0],[3,4]]
```

# 3 Selecting a range of data with a slice 

```python
# slice column labels
players=['L. Messi','Cristiano Ronaldo']
df.iloc[[0,1],0:]
```

# 4. Selecting with a range of data with a slice

```python
# one condition: Select players with height above 180 cm
column=[1,2,3,4]
df.iloc[list(df['height_cm']>180),column]
```

```python
# multiple conditions : select player with height above 180cm
boollist=list((df['height_cm']>180)&(df['nationality']=='Argentina'))
df.iloc[boollist,column]
```
