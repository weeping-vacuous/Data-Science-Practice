```python
import pandas as pd
```

```python
df_players=pd.read_csv('players_20.csv')
```

```python
df_players.set_index('short_name',inplace=True)
```

```python
df_players.head(4)
```

```python
df=df_players[['long_name'	,'age',	'dob',	'height_cm'	,'weight_kg',	'nationality',	'club']]
```

```python
#weight of chritiaono ronaldo
df.loc['Cristiano Ronaldo','weight_kg']
```

```python
df.loc[:,'height_cm']
```

# 2 Selecting with a list of values

```python
# get all data about L.Messi and christiano ronaldo and l.messi
df.loc[['L. Messi','Cristiano Ronaldo']]
```

# 4. Selecting with a range of data with a slice

```python
# one condition: Select players with height above 180 cm
column=['age','dob','height_cm','weight_kg']
df.loc[df['height_cm']>180,column]
```

```python
# multiple conditions: Select player with height above 180cm from Argentin
df.loc[(df['height_cm']>180)&(df['nationality']=='Argentina'),column]
```
