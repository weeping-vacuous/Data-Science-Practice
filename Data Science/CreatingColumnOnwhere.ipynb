# Creating a column based on 1 condition

```python
import numpy as np
```

```python
# create an array based on price tiers
np.where(df_laptops['Price_euros']>2000,'Expensive','Cheap')
```

```python
df_laptops['Price_tier']=np.where(df_laptops['Price_euros']>2000,'Expensive','Cheap')
df_laptops.head(5)
```

```python
df_laptops.value_counts('Price_tier')
```

# Creating a column for more than 2 choices using np.select

```python
# lets reset df first
df_laptops=df_laptops.drop('Screen_tiers',axis=1)

df_laptops.drop('Price_tier',axis=1,inplace=True)
df_laptops
```

```python
#create an array based on multiples conditions

# create conditions and value variables first
conditions=[
     df_laptops['Price_euros']>3000,
     (df_laptops['Price_euros']>2000)&(df_laptops['Price_euros']<=3000),
     (df_laptops['Price_euros']>800)&(df_laptops['Price_euros']<=2000)
]
values=['Too Expensive','Expensive','Affordable']
```

```python
# set it to a new column
np.select(conditions, values,'Cheap')
```

```python
df_laptops['Price_tiers']=np.select(conditions, values,'Cheap')
df_laptops
```

# Exercise

```python
# screen tiers
conditions=[
           df_laptops['Inches']>16,
           (df_laptops['Inches']>14)&(df_laptops['Inches']<=16),
           (df_laptops['Inches']>12)&(df_laptops['Inches']<=14),
           df_laptops['Inches']<=12,
]

values=['Too Big','Big','Small','Too Small']
```

```python
df_laptops['Screen_tiers']=np.select(conditions,values,default='NaN')
df_laptops
```

```python
df_laptops.value_counts('Screen_tiers')
```
